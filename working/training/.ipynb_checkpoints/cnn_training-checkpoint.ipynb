```python

import os
import warnings
import itertools
import numpy as np
import pandas as pd
import librosa
import matplotlib.pyplot as plt
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchmetrics.classification import MulticlassF1Score

from sklearn.metrics import classification_report, confusion_matrix

warnings.filterwarnings("ignore", category=UserWarning)
from google.colab import drive
drive.mount('/content/drive')


class CONFIG:
    PROCESSED_DIR = "/content/drive/MyDrive/Speech_Emotion_Recognition/working/processed"

    #đường dẫn model
    MODEL_DIR = "/content/drive/MyDrive/Speech_Emotion_Recognition/working/models"
    MODEL_OUT = os.path.join(MODEL_DIR, "ravdess_ser_cnn_pytorch.pth")

    #tham số trích xuất đặc trưng
    SR_FEAT = 16000
    N_MELS = 128
    MAX_PAD_LEN = 200  #3.2 giây

    #tham số huấn luyện
    BATCH_SIZE = 32
    EPOCHS = 100
    NUM_CLASSES = 7
    INPUT_SHAPE = (1, N_MELS, MAX_PAD_LEN) # (Channels, Height, Width)

    # Mapping
    EMOTION_MAP = {
        0: 'neutral', # neutral + calm
        1: 'happy',
        2: 'sad',
        3: 'angry',
        4: 'fearful',
        5: 'disgust',
        6: 'surprised'
    }
    ID_TO_NAME_MAP = EMOTION_MAP

# Tạo thư mục lưu model nếu chưa có
os.makedirs(CONFIG.MODEL_DIR, exist_ok=True)


#Các hàm trích xuất đặc trưng
def add_white_noise(y, noise_factor=0.002):
    """Thêm nhiễu trắng nhẹ"""
    noise = np.random.randn(len(y))
    return y + noise_factor * noise

def pitch_shift(y, sr=CONFIG.SR_FEAT, n_steps=2):
    """Thay đổi cao độ ngẫu nhiên"""
    n = np.random.uniform(-n_steps, n_steps)
    return librosa.effects.pitch_shift(y, sr=sr, n_steps=n)

def extract_features(y, sr=CONFIG.SR_FEAT, n_mels=CONFIG.N_MELS, max_pad_len=CONFIG.MAX_PAD_LEN):
    """Trích xuất LogMel Spectrogram, chuẩn hóa và pad"""
    try:
        melspec = librosa.feature.melspectrogram(
            y=y, sr=sr, n_mels=n_mels, n_fft=2048, hop_length=512, fmax=8000
        )
        log_melspec = librosa.power_to_db(melspec, ref=np.max)

        # Normalize
        log_melspec = (log_melspec - log_melspec.mean()) / (log_melspec.std() + 1e-8)

        # Pad or truncate
        if log_melspec.shape[1] < max_pad_len:
            pad_width = max_pad_len - log_melspec.shape[1]
            log_melspec = np.pad(log_melspec, pad_width=((0, 0), (0, pad_width)), mode='constant')
        else:
            log_melspec = log_melspec[:, :max_pad_len]

        return log_melspec
    except Exception as e:
        print(f"Lỗi khi trích xuất đặc trưng: {e}")
        return None

#PyTorch Data Generator
class AudioDataset(Dataset):
    def __init__(self, df, base_dir, augment=False):
        self.df = df
        self.base_dir = base_dir
        self.paths = df['file_path'].values
        self.labels = df['label_id'].values
        self.augment = augment

    def __len__(self):
        return len(self.df)

    def __getitem__(self, index):
        # Lấy đường dẫn tương đối từ dataframe
        relative_path = self.paths[index]
        label = self.labels[index]

        file_path = os.path.join(self.base_dir, relative_path)
        try:
            y_audio, _ = librosa.load(file_path, sr=CONFIG.SR_FEAT)

            # Augmentation
            if self.augment and np.random.rand() < 0.5:
                if np.random.rand() < 0.5:
                    y_audio = add_white_noise(y_audio)
                else:
                    y_audio = pitch_shift(y_audio)

            feat = extract_features(y_audio)
            if feat is None:
                # Trả về mảng zero nếu có lỗi
                feat = np.zeros((CONFIG.N_MELS, CONFIG.MAX_PAD_LEN))

        except Exception as e:
            print(f"Lỗi khi tải file {file_path}: {e}")
            feat = np.zeros((CONFIG.N_MELS, CONFIG.MAX_PAD_LEN))

        # Thêm chiều channel và chuyển sang Tensor
        features_tensor = torch.from_numpy(feat).float().unsqueeze(0) # Shape: (1, N_MELS, MAX_PAD_LEN)
        label_tensor = torch.tensor(label, dtype=torch.long)

        return features_tensor, label_tensor

#Kiến trúc mô hình CNN
class SimpleCNN(nn.Module):
    def __init__(self, num_classes=CONFIG.NUM_CLASSES):
        super(SimpleCNN, self).__init__()

        # Block 1
        self.block1 = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding='same'),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Conv2d(32, 32, kernel_size=3, padding='same'),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.25)
        )

        # Block 2
        self.block2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, padding='same'),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, kernel_size=3, padding='same'),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.25)
        )

        # Block 3
        self.block3 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding='same'),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 128, kernel_size=3, padding='same'),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.25)
        )

        # Global Pooling and Dense layers
        self.global_pool = nn.AdaptiveAvgPool2d((1, 1))
        self.dense_block = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128, 256),
            nn.ReLU(),
            nn.BatchNorm1d(256),
            nn.Dropout(0.5),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        x = self.global_pool(x)
        x = self.dense_block(x)
        return x

#hàm hỗ trợ
def plot_history(history):
    fig, axes = plt.subplots(1, 3, figsize=(21, 5))

    # Loss
    axes[0].plot(history['train_loss'], label='Train Loss')
    axes[0].plot(history['val_loss'], label='Val Loss')
    axes[0].set_title('Training and Validation Loss')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # Accuracy
    axes[1].plot(history['train_acc'], label='Train Acc')
    axes[1].plot(history['val_acc'], label='Val Acc')
    axes[1].set_title('Training and Validation Accuracy')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Accuracy')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    # F1 Score
    axes[2].plot(history['train_f1'], label='Train F1 (Macro)')
    axes[2].plot(history['val_f1'], label='Val F1 (Macro)')
    axes[2].set_title('Training and Validation F1-Score')
    axes[2].set_xlabel('Epoch')
    axes[2].set_ylabel('F1-Score')
    axes[2].legend()
    axes[2].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(cm, classes):
    plt.figure(figsize=(10, 8))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix', fontsize=14)
    plt.colorbar()
    ticks = np.arange(len(classes))
    plt.xticks(ticks, classes, rotation=45, ha='right')
    plt.yticks(ticks, classes)

    thresh = cm.max() / 2.0
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black", fontsize=10)

    plt.ylabel('True label', fontsize=12)
    plt.xlabel('Predicted label', fontsize=12)
    plt.tight_layout()
    plt.show()

#Pipeline Huấn luyện

def run_training():
    # Thiết lập
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Sử dụng thiết bị: {device}")

    # Add this line to enable synchronous CUDA error reporting for debugging
    os.environ['CUDA_LAUNCH_BLOCKING'] = '1'

    # Tải dữ liệu
    try:
        train_csv_path = os.path.join(CONFIG.PROCESSED_DIR, "train_final.csv")
        val_csv_path = os.path.join(CONFIG.PROCESSED_DIR, "val_final.csv")
        test_csv_path = os.path.join(CONFIG.PROCESSED_DIR, "test_final.csv")

        train_df = pd.read_csv(train_csv_path)
        val_df = pd.read_csv(val_csv_path)
        test_df = pd.read_csv(test_csv_path)
    except FileNotFoundError as e:
        print(f"Lỗi: Không tìm thấy file CSV tại '{CONFIG.OUTPUT_DIR}'")
        print("Hãy đảm bảo biến OUTPUT_DIR được cấu hình đúng và bạn đã chạy file tiền xử lý")
        print(e)
        return

    print("="*60)
    print(f"Dataset:")
    print(f"   Train: {len(train_df)} | Val: {len(val_df)} | Test: {len(test_df)}")
    print("="*60)

    # Tạo DataLoaders
    train_dataset = AudioDataset(train_df, base_dir=CONFIG.PROCESSED_DIR, augment=True)
    val_dataset = AudioDataset(val_df,base_dir=CONFIG.PROCESSED_DIR, augment=False)

    train_loader = DataLoader(train_dataset, batch_size=CONFIG.BATCH_SIZE, shuffle=True, num_workers=4, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=CONFIG.BATCH_SIZE, shuffle=False, num_workers=4, pin_memory=True)

    # Khởi tạo mô hình, loss, optimizer, scheduler
    model = SimpleCNN().to(device)
    print(model)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=30, T_mult=1, eta_min=1e-6)

    # Khởi tạo đối tượng tính F1-Score
    train_f1_metric = MulticlassF1Score(num_classes=CONFIG.NUM_CLASSES, average='macro').to(device)
    val_f1_metric = MulticlassF1Score(num_classes=CONFIG.NUM_CLASSES, average='macro').to(device)

    # Biến cho Early Stopping & history
    best_val_f1 = 0.0
    patience_counter = 0
    patience = 20
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': [], 'train_f1': [], 'val_f1': []}

    # --- Vòng lặp huấn luyện ---
    for epoch in range(CONFIG.EPOCHS):
        # --- Training Phase ---
        model.train()
        running_loss, correct_preds, total_preds = 0.0, 0, 0
        train_f1_metric.reset()

        train_pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{CONFIG.EPOCHS} [Train]")
        for inputs, labels in train_pbar:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            _, preds = torch.max(outputs, 1)

            running_loss += loss.item() * inputs.size(0)
            correct_preds += torch.sum(preds == labels.data)
            total_preds += inputs.size(0)
            train_f1_metric.update(preds, labels)

            train_pbar.set_postfix({'loss': loss.item()})

        epoch_train_loss = running_loss / total_preds
        epoch_train_acc = (correct_preds.double() / total_preds).item()
        epoch_train_f1 = train_f1_metric.compute().item()

        # --- Validation Phase ---
        model.eval()
        val_loss, val_correct, val_total = 0.0, 0, 0
        val_f1_metric.reset()

        val_pbar = tqdm(val_loader, desc=f"Epoch {epoch+1}/{CONFIG.EPOCHS} [Val]")
        with torch.no_grad():
            for inputs, labels in val_pbar:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                _, preds = torch.max(outputs, 1)

                val_loss += loss.item() * inputs.size(0)
                val_correct += torch.sum(preds == labels.data)
                val_total += inputs.size(0)
                val_f1_metric.update(preds, labels)

                val_pbar.set_postfix({'loss': loss.item()})

        epoch_val_loss = val_loss / val_total
        epoch_val_acc = (val_correct.double() / val_total).item()
        epoch_val_f1 = val_f1_metric.compute().item()

        # --- Lưu lịch sử và cập nhật scheduler ---
        history['train_loss'].append(epoch_train_loss)
        history['train_acc'].append(epoch_train_acc)
        history['train_f1'].append(epoch_train_f1)
        history['val_loss'].append(epoch_val_loss)
        history['val_acc'].append(epoch_val_acc)
        history['val_f1'].append(epoch_val_f1)
        scheduler.step()

        print(f"Epoch {epoch+1}: Train Loss: {epoch_train_loss:.4f}, Acc: {epoch_train_acc:.4f}, F1: {epoch_train_f1:.4f} | Val Loss: {epoch_val_loss:.4f}, Acc: {epoch_val_acc:.4f}, F1: {epoch_val_f1:.4f}")

        # --- Early Stopping & Model Checkpoint (dựa trên val_f1) ---
        if epoch_val_f1 > best_val_f1:
            best_val_f1 = epoch_val_f1
            torch.save(model.state_dict(), CONFIG.MODEL_OUT)
            print(f"Validation F1-score improved to {best_val_f1:.4f}. Saving model to {CONFIG.MODEL_OUT}")
            patience_counter = 0
        else:
            patience_counter += 1
            print(f"Validation F1-score did not improve. Patience: {patience_counter}/{patience}")

        if patience_counter >= patience:
            print("Early stopping triggered")
            break
    print("\n" + "="*50)
    print("Lịch sử quá trình huấn luyện:")
    plot_history(history)
    print("="*50)
if __name__ == "__main__":
    run_training()
```

```python
import itertools
import matplotlib.pyplot as plt
import numpy as np

def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion Matrix'):
    if normalize:
        # Thêm +1e-8 để tránh lỗi chia cho 0 nếu một hàng toàn là 0
        cm = cm.astype('float') / (cm.sum(axis=1)[:, np.newaxis] + 1e-8)
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.figure(figsize=(10, 8))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(title, fontsize=14)
    plt.colorbar()
    ticks = np.arange(len(classes))
    plt.xticks(ticks, classes, rotation=45, ha='right')
    plt.yticks(ticks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black", fontsize=10)

    plt.ylabel('True label', fontsize=12)
    plt.xlabel('Predicted label', fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
test_csv_path = os.path.join(CONFIG.PROCESSED_DIR, "test_final.csv")
test_df = pd.read_csv(test_csv_path)
print(" Bắt đầu quá trình đánh giá trên tập Test...")

#Thiết lập và tải lại dữ liệu Test
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Sử dụng thiết bị: {device}")

try:
    test_csv_path = os.path.join(CONFIG.PROCESSED_DIR, "test_final.csv")
    test_df = pd.read_csv(test_csv_path)
    print(f"Đã tải thành công test_final.csv với {len(test_df)} mẫu.")
except FileNotFoundError:
    print(f"Lỗi: Không tìm thấy file 'test_final.csv' tại '{CONFIG.PROCESSED_DIR}'")

test_dataset = AudioDataset(test_df, base_dir=CONFIG.PROCESSED_DIR, augment=False)
test_loader = DataLoader(test_dataset, batch_size=CONFIG.BATCH_SIZE, shuffle=False, num_workers=2)

# Tải lại Model đã huấn luyện
evaluation_model = SimpleCNN(num_classes=CONFIG.NUM_CLASSES).to(device)
try:
    # Added map_location to load the model on CPU if CUDA is not available
    evaluation_model.load_state_dict(torch.load(CONFIG.MODEL_OUT, map_location=torch.device('cpu')))
    print(f"Đã tải thành công trọng số từ: {CONFIG.MODEL_OUT}")
except FileNotFoundError:
    print(f"Lỗi: Không tìm thấy file model '{CONFIG.MODEL_OUT}'. Hãy đảm bảo cell huấn luyện đã chạy thành công.")
evaluation_model.eval()

#Thực hiện dự đoán trên tập Test
print("Thực hiện dự đoán trên tập Test...")

y_true = []
y_pred = []

with torch.no_grad():
    for inputs, labels in tqdm(test_loader, desc="[Test Predictions]"):
        inputs = inputs.to(device)
        outputs = evaluation_model(inputs)
        _, preds = torch.max(outputs, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(preds.cpu().numpy())

print("Hoàn tất dự đoán")


from sklearn.metrics import accuracy_score, f1_score

# Chuyển thành mảng numpy để tính toán
y_true_np = np.array(y_true)
y_pred_np = np.array(y_pred)

# Tính toán các chỉ số
test_accuracy = accuracy_score(y_true_np, y_pred_np)
test_f1_macro = f1_score(y_true_np, y_pred_np, average='macro', zero_division=0)
test_f1_weighted = f1_score(y_true_np, y_pred_np, average='weighted', zero_division=0)

print("\n" + "="*60)
print(" Kết quả tổng thể trên tập Test:")
print("="*60)
print(f"   - Accuracy:   {test_accuracy * 100:.2f}%")
print(f"   - F1-Score (Macro):   {test_f1_macro:.4f}")
print(f"   - F1-Score (Weighted): {test_f1_weighted:.4f}")
print("="*60)


# Hiển thị classification report chi tiết và ma trận nhập nhằng
expected_labels = list(range(CONFIG.NUM_CLASSES))
target_names = [CONFIG.ID_TO_NAME_MAP[i] for i in expected_labels]

print("\n" + "="*60)
print(" Báo cáo Phân loại chi tiết (Classification Report):")
print("="*60)
print(classification_report(
    y_true,
    y_pred,
    labels=expected_labels,
    target_names=target_names,
    digits=4,
    zero_division=0
))
print("="*60)

print("\n" + "="*60)
print(" Ma trận nhập nhằng (Confusion Matrix):")
print("="*60)

cm = confusion_matrix(y_true, y_pred, labels=expected_labels)

plot_confusion_matrix(cm, classes=target_names, normalize=False,
                     title="Confusion Matrix (Counts)")

plot_confusion_matrix(cm, classes=target_names, normalize=True,
                     title="Normalized Confusion Matrix")

print("Hoàn tất quá trình đánh giá.")
```
